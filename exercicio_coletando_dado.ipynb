01. EXECEL para CSV

```python
from openpyxl import load_workbook
import csv

#carregando o arquivo Excel
planilhas = load_workbook(filename = 'credito.xlsx')
planilha = planilhas.active

#Criando listas
default = []
estado_civil = []
id = []
sexo = []
idade = []

#Extraindo as colunas
cabecalho = next(planilha.values)
indice_default = cabecalho.index('default')
indice_estado_civil = cabecalho.index('estado_civil')
indice_id = cabecalho.index('id')
indice_sexo = cabecalho.index('sexo')
indice_idade = cabecalho.index('idade')
```

```python
#Filtrando clientes inadimplentes e solteiros
filtro_id = [linha[indice_id] for linha in planilha.values if linha[indice_default] == 1 and linha [indice_estado_civil]== 'solteiro']
filtro_idade = [linha[indice_idade] for linha in planilha.values if linha[indice_default] == 1 and linha [indice_estado_civil]== 'solteiro']
filtro_sexo = [linha[indice_sexo] for linha in planilha.values if linha[indice_default] == 1 and linha [indice_estado_civil]== 'solteiro']
print(filtro_id)
print(filtro_idade)
print(filtro_sexo)
```

```python
#Criando um arquivo CSV para escrita 'credito.csv'
with open (file = 'credito.csv', mode = 'w',encoding ='utf8') as csv_arquivo:
  writer_csv = csv.writer(csv_arquivo, delimiter = ';')
  #criando um cabecalho
  writer_csv.writerow(['id','sexo','idade'])

  for id, sexo, idade in zip(filtro_id,filtro_sexo, filtro_idade):
    writer_csv.writerow([id,sexo,idade])
```
